## open()

파일은 open() 명령에서 `<MODE>` 값에 따라 읽고 쓴다. [Python builtin functions](https://docs.python.org/3/library/functions.html)
    
>open(file, mode='rb', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
  - mode는 파일에 대한 액션과 파일 형식을 지정한다: rb, wb, rt, wt
     - 액션 : w(기록), a(추가), r(읽기:기본값)
     - 타입 : t(텍스트:기본값), b(2진 파일)
open 으로 열린 객체는 `.close()` 로 닫아 주어야 한다.

file:
 - Text file : 문자코드(ascii code)
 - Binary file : byte streams

`readline()` 함수 이용하기

```python
f = open("1-05Sequence_0.ipynb", "rt", encoding="utf8")
```

```python
f.mode, f.name, f.encoding
```

```python
while True:
    line = f.readline(10) # ->(한줄씩)EOL \n \r\n EOF
    if not line:
        break
    line = line.strip() # 양쪽 끝에 있는 공백 문자 제거
    print(line)

f.close()
```

### 내용 추가하기

`append` 모드는 파일이 있으면 뒤에 내용을 추가, `<MODE>`를 `a`로 제시하면 파일의 끝에 내용을 추가할 수 있다.



```python
f = open('testfile.txt', 'a')
for i in range(11, 20):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

### with ~ as
`with ~ as` 를 이용하면 파일 close 작업을 자동으로 수행 -> 안전하게 시스템 자원을 활용
즉 close() 메서드를 명시하지 않아도, 파일을 다 사용후에 닫아 준다.

`split()` 를 사용해 파일을 불러와서 개행문자 구분한 리스트로 반환

```python
with open("1-05Sequence_0.ipynb", "rt", encoding="utf8") as f:
    lines = f.readlines(:-1)
len(lines)
```

```python
type(lines)
```

```python
lines[:10]
```

```python
# 루프를 돌려 출력
for line in lines:
    print(line.strip())
```

#### 문제3

현재 사용하는 1-05Sequence_0.ipynb 파일을 읽어 파일에 있는 단어의 개수를 세는 프로그램을 작성하시오.

```python
with open ("1-05Sequence_0.ipynb", "rt", encoding="utf8") as f:
    line = f.readlines()
```

```python
# 파일 읽기
# readlines() 모두 읽기.
# 리스트 크기 만큼 반복
   # 문자열을 ' ' 분리 split()
   # 결과를 len() = ++sum

with open ("1-05Sequence_0.ipynb", "rt", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        data = line.split(' ')
          
print(len(data))
f.close()
```

```python
result = []
with open("1-05Sequence_0.ipynb", "rt", encoding="utf-8") as f:
    for line in f.readlines():
        word_list = list(line)
        # print(word_list)
        for _ in range(len(word_list)):
            result.append(word_list.pop())
            
print("파일의 단어 수 {}개".format(len(result)))
```

```python
result = 0
with open("1-05Sequence_0.ipynb", mode = 'r', encoding = 'utf-8')as f:
    for line in f.readlines():
        word_list = "".join(line).split()
        result += len(word_list)
print("파일의 단어 수 {}개".format(result))
```

## binary mode

바이너리 파일은 바이너리 모드로 작성해서 사용해야 한다. 

```python
"""
바이너리 파일 생성
"""

print("====== binary file =====")
f = open("test.bin", "wb") f
    
    for i in range(0, 256):
    data = i.to_bytes(1, 'little')
    f.write(data)
f.close()
```

```python
# with
print("====== with binary file =====")
with open("test.bin", "wb") as f:     # wb -> binary 파일을 읽을 때 

    for i in range(0, 256):
        data = i.to_bytes(1, 'little')
        f.write(data)
```

#### 바이너리 파일 복사

바이너리 모드로 rose-flower.jpeg를 읽어 들여서, 바이너리 모드로 rose-flower-copy.jpeg로 저장해 보자.

```python
with open("rose-flower.jpeg", "rb") as f:
    data = f.read()

print("읽은 데이터:", data)

# 저장
with open("rose-flower-copy.jpeg", "wb") as f:
    f.write(data)
```
